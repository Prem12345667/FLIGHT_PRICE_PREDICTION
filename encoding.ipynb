```python
encoded_airline = pd.get_dummies(df["airline"])
encoded_airline
```

```python
df = pd.concat([df,encoded_airline],axis=1)
df[["AirAsia","Air_India","GO_FIRST","Indigo","SpiceJet","Vistara"]]=df[["AirAsia","Air_India","GO_FIRST","Indigo","SpiceJet","Vistara"]].astype("int")
df
```

```python
df["airline"].isnull().sum()
```

```python
source_enc = pd.get_dummies(df["source_city"],prefix="source")
source_enc
```

```python
df = pd.concat([df,source_enc],axis=1)
df[["source_Bangalore","source_Chennai","source_Delhi","source_Hyderabad","source_Kolkata","source_Mumbai"]] = df[["source_Bangalore","source_Chennai","source_Delhi","source_Hyderabad","source_Kolkata","source_Mumbai"]].astype("int")
df
```

```python
destination_enc = pd.get_dummies(df["destination_city"],prefix="dest")
destination_enc
```

```python
df = pd.concat([df,destination_enc],axis=1)
df[["dest_Bangalore","dest_Chennai","dest_Delhi","dest_Hyderabad","dest_Kolkata","dest_Mumbai"]] = df[["dest_Bangalore","dest_Chennai","dest_Delhi","dest_Hyderabad","dest_Kolkata","dest_Mumbai"]].astype("int")
df
```

```python
dep_tim = pd.get_dummies(df["departure_time"],prefix="deptime")
dep_tim
df = pd.concat([df,dep_tim],axis=1)
df
```

```python
dep_tim = pd.get_dummies(df["departure_time"],prefix="deptime")
dep_tim
```

```python

df[["deptime_Afternoon","deptime_Early_Morning","deptime_Evening","deptime_Late_Night","deptime_Morning","deptime_Night"]] = df[["deptime_Afternoon","deptime_Early_Morning","deptime_Evening","deptime_Late_Night","deptime_Morning","deptime_Night"]].astype("int")
df
```

```python
arr_tim = pd.get_dummies(df["arrival_time"],prefix="arrtime",)
arr_tim
```

```python
df = pd.concat([df,arr_tim],axis=1)
df[["arrtime_Afternoon","arrtime_Early_Morning"	,"arrtime_Evening","arrtime_Late_Night","arrtime_Morning","arrtime_Night"]]= df[["arrtime_Afternoon","arrtime_Early_Morning","arrtime_Evening","arrtime_Late_Night","arrtime_Morning","arrtime_Night"]].astype("int")
df
```

```python
df["stops"] = df["stops"].replace(["zero", "one", "two_or_more"], [0, 1, 2])
df.head()
```

```python
df["class"].value_counts()
```

```python
df["class"] = df["class"].replace(["Economy","Business"],[0,1])
df.head()
```

```python
df.drop(df[["airline","flight","source_city","departure_time","arrival_time","destination_city","index"]],axis=1,inplace=True)
```
